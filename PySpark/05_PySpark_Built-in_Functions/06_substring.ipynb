## ğŸ“¦ ×™×™×‘×•× ×¡×¤×¨×™×•×ª ×•×”×›× ×ª × ×ª×•× ×™×

×¨××©×™×ª, × ×™×™×‘× ××ª ×”×¡×¤×¨×™×•×ª ×”× ×“×¨×©×•×ª ×•× ×™×¦×•×¨ DataFrame ×œ×“×•×’××”:

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring

spark = SparkSession.builder.appName("stringoperations").getOrCreate()

# ×™×¦×™×¨×ª DataFrame ×¢× ×ª××¨×™×›×™× ×‘×¤×•×¨××˜ "year month day"
data = [
    ("20200828",),
    ("20180525",),
]

columns = ["id", "date"]
df = spark.createDataFrame(data, ["date"])
df.printSchema()
df.show(truncate=False)
````

## ğŸ’¡ ×“×•×’××” 1: ×©×™××•×© ×‘-substring() ×¢× withColumn() <a id='example1'></a>

×”×“×•×’××” ×”×‘××” ××“×’×™××” ×©×™××•×© ×‘-`substring()` ×¢× `withColumn()`.

×›××Ÿ ×× ×—× ×• ××—×œ×¦×™× ××ª ×”×©× ×”, ×”×—×•×“×© ×•×”×™×•× ×›×¢××•×“×•×ª × ×¤×¨×“×•×ª:

````python
# ×—×™×œ×•×¥ ×©× ×”, ×—×•×“×©, ×™×•×
df2 = ((df.withColumn("year", substring("date", 1, 4))
       .withColumn("month", substring("date", 5, 2)))
       .withColumn("day", substring("date", 7, 2)))

df2.printSchema()
df2.show()
````

### ğŸ“‹ ×”×¡×‘×¨ ×”×§×•×“:

- **year**: ××—×œ×¥ 4 ×ª×•×•×™× ×¨××©×•× ×™× (×ª×•×•×™× 1-4)
- **month**: ××—×œ×¥ 2 ×ª×•×•×™× ×”×—×œ ×××™×§×•× 5 (×ª×•×•×™× 5-6)
- **day**: ××—×œ×¥ 2 ×ª×•×•×™× ×”×—×œ ×××™×§×•× 7 (×ª×•×•×™× 7-8)

## ğŸ” ×“×•×’××” 2: substring() ×¢× select() <a id='example2'></a>

××¤×©×¨ ×œ×”×©×ª××© ×‘-`substring()` ×¢× `select()` ×›×“×™ ×œ×§×‘×œ ××•×ª×” ×ª×•×¦××”:

````python
# ×©×™××•×© ×¢× select()
df.select(
    "date",
    substring("date", 1, 4).alias("year"),
    substring("date", 5, 2).alias("month"),
    substring("date", 7, 2).alias("day")
).show()
````

## ğŸ“Š ×“×•×’××” 3: substring() ×¢× selectExpr() <a id='example3'></a>

×“×•×’××” ×œ×©×™××•×© ×‘-`selectExpr()` ×›×“×™ ×œ×§×‘×œ ×ª×ª-××—×¨×•×–×ª ×©×œ ×¢××•×“×”(date):

````python
# ×©×™××•×© ×¢× selectExpr()
df2 = df.selectExpr(
    "date",
    "substring(date, 1, 4) as year",
    "substring(date, 5, 2) as month",
    "substring(date, 7, 2) as day"
)

df2.show()
````

## ğŸ”§ ×“×•×’××” 4: ×©×™××•×© ×‘-substr() ×-Column type <a id='example4'></a>

×œ×”×œ×Ÿ ×”×“×•×’××” ×œ×—×™×œ×•×¥ ×ª×ª-××—×¨×•×–×ª ×‘×××¦×¢×•×ª ×¤×•× ×§×¦×™×” `substr()` ××”×˜×™×¤×•×¡ `pyspark.sql.Column`:

````python
# ×©×™××•×© ×‘-substr ×-Column type
df3 = df.withColumn("year", col("date").substr(1, 4)) \
        .withColumn("month", col("date").substr(5, 2)) \
        .withColumn("day", col("date").substr(7, 2))

df3.show()
````

## ğŸ“ ×“×•×’××” ××œ××” <a id='complete-example'></a>

×”× ×” ×“×•×’××” ××§×™×¤×” ×”××©×œ×‘×ª ××ª ×›×œ ××” ×©×œ××“× ×•:

````python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring

spark = SparkSession.builder.appName("stringoperations").getOrCreate()

data = [("20200828",), ("20180525",)]
columns = ["id", "date"]
df = spark.createDataFrame(data, ["date"])
df.printSchema()
df.show(truncate=False)

# ×©×™××•×© ×‘-substring() ×¢× withColumn
df.withColumn("year", substring("date", 1, 4)) \
  .withColumn("month", substring("date", 5, 2)) \
  .withColumn("day", substring("date", 7, 2)) \
  .printSchema()

df.withColumn("year", substring("date", 1, 4)) \
  .withColumn("month", substring("date", 5, 2)) \
  .withColumn("day", substring("date", 7, 2)) \
  .show()

# ×©×™××•×© ×¢× select
df1 = df.select(
    "date",
    substring("date", 1, 4).alias("year"),
    substring("date", 5, 2).alias("month"),
    substring("date", 7, 2).alias("day")
)

# ×©×™××•×© ×¢× selectExpr
df2 = df.selectExpr(
    "date",
    "substring(date, 1, 4) as year",
    "substring(date, 5, 2) as month",
    "substring(date, 7, 2) as day"
)

# ×©×™××•×© ×‘-substr ×-Column type
df3 = df.withColumn("year", col("date").substr(1, 4)) \
        .withColumn("month", col("date").substr(5, 2)) \
        .withColumn("day", col("date").substr(7, 2))

df3.show()
````
