## ğŸ“ ×™×¦×™×¨×ª DataFrame ×œ×“×•×’××”

× ×™×¦×•×¨ DataFrame ×¤×©×•×˜ ×¢× × ×ª×•× ×™ ×¢×•×‘×“×™× ×©×™×©××© ××•×ª× ×• ×œ×”×“×’××ª ×¤×•× ×§×¦×™×•×ª ×”××’×¨×’×¦×™×”:

```python
# ×”×’×“×¨×ª ×”× ×ª×•× ×™×
simpleData = [
    ("James", "Sales", 3000),
    ("Michael", "Sales", 4600),
    ("Robert", "Sales", 4100),
    ("Maria", "Finance", 3000),
    ("James", "Sales", 3000),
    ("Scott", "Finance", 3300),
    ("Jen", "Finance", 3900),
    ("Jeff", "Marketing", 3000),
    ("Kumar", "Marketing", 2000),
    ("Saif", "Sales", 4100)
]

# ×”×’×“×¨×ª ×”×¡×›×™××”
schema = ["employee_name", "department", "salary"]

# ×™×¦×™×¨×ª DataFrame
df = spark.createDataFrame(data=simpleData, schema=schema)
```

## 1ï¸âƒ£ approx_count_distinct() - ×¡×¤×™×¨×” ××©×•×¢×¨×ª ×©×œ ×¢×¨×›×™× ×™×™×—×•×“×™×™×

## 4ï¸âƒ£ collect_set() - ××™×¡×•×£ ×¢×¨×›×™× ×™×™×—×•×“×™×™× ×›×¡×˜

×”×¤×•× ×§×¦×™×” ××—×–×™×¨×” ××ª ×›×œ ×”×¢×¨×›×™× ××¢××•×“×ª ×”×§×œ×˜ **×œ×œ× ×¢×¨×›×™× ×›×¤×•×œ×™×**.

## 5ï¸âƒ£ countDistinct() - ×¡×¤×™×¨×ª ×¢×¨×›×™× ×™×™×—×•×“×™×™×

×”×¤×•× ×§×¦×™×” ××—×–×™×¨×” ××ª ××¡×¤×¨ ×”××œ×× ×˜×™× ×”×™×™×—×•×“×™×™× ×‘×¢××•×“×•×ª. × ×™×ª×Ÿ ×œ×”×¢×‘×™×¨ ××¡×¤×¨ ×¢××•×“×•×ª.
