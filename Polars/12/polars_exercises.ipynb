### âœ… ×¤×ª×¨×•×Ÿ

````python
# ×¤×ª×¨×•×Ÿ ××œ×:

# ×©×œ×‘ 1: ×”×’×“×¨×ª ×¤×•× ×§×¦×™×•×ª ××•×“×•×œ×¨×™×•×ª
def filter_water_or_electric(lf: pl.LazyFrame) -> pl.LazyFrame:
    """
    ××¡× ×Ÿ ×¨×§ ×¤×•×§×™××•× ×™× ××¡×•×’ Water ××• Electric
    """
    accepted_types = ['Water', 'Electric']
    return (
        lf
        .filter(
            (pl.col('Type 1').is_in(accepted_types)) |
            (pl.col('Type 2').is_in(accepted_types))
        )
    )

def add_power_ratio(lf: pl.LazyFrame) -> pl.LazyFrame:
    """
    ××•×¡×™×£ ×¢××•×“×ª power_ratio = Attack / Defense
    """
    return (
        lf
        .with_columns(
            power_ratio=(pl.col('Attack') / pl.col('Defense'))
        )
    )

def filter_by_ratio(lf: pl.LazyFrame, threshold: float = 0.8) -> pl.LazyFrame:
    """
    ××¡× ×Ÿ ×œ×¤×™ power_ratio
    """
    return lf.filter(pl.col('power_ratio') > threshold)

def select_relevant_cols(lf: pl.LazyFrame) -> pl.LazyFrame:
    """
    ×‘×•×—×¨ ×¨×§ ×¢××•×“×•×ª ×¨×œ×•×•× ×˜×™×•×ª
    """
    return lf.select(
        'Name',
        'Type 1',
        'Type 2',
        'Total',
        'Attack',
        'Defense',
        'power_ratio'
    )

# ×©×œ×‘ 2: ×‘× ×™×™×ª ×”-pipeline
print("ğŸ” ×‘×•×“×§ ××ª Query Plan:\n")
query_plan = (
    lf
    .pipe(filter_water_or_electric)  # 1. ×¡×™× ×•×Ÿ ×œ×¤×™ ×˜×™×¤×•×¡
    .pipe(add_power_ratio)           # 2. ×”×•×¡×¤×ª power_ratio
    .pipe(filter_by_ratio, 0.8)      # 3. ×¡×™× ×•×Ÿ ×œ×¤×™ ratio
    .pipe(select_relevant_cols)      # 4. ×‘×—×™×¨×ª ×¢××•×“×•×ª
    .sort('Total', descending=True)  # 5. ××™×•×Ÿ
    .head(10)                        # 6. ×œ×§×™×—×ª 10 ×¨××©×•× ×™×
)

print(query_plan.explain())
print("\n" + "="*50 + "\n")

# ×©×œ×‘ 3: ×”×¨×¦×”
print("ğŸ“Š ×ª×•×¦××•×ª:\n")
result = query_plan.collect()
print(result)

# ×©×œ×‘ 4: ×‘×“×™×§×ª ××™×›×•×ª
print("\n" + "="*50 + "\n")
print("âœ… ×‘×“×™×§×ª ××™×›×•×ª:\n")
check = Check(CheckLevel.WARNING, 'Power Ratio Quality')
quality_check = (
    check
    .is_greater_than('power_ratio', 0)  # ×›×œ ×”×¢×¨×›×™× ×—×™×•×‘×™×™×?
    .is_complete('power_ratio')         # ××™×Ÿ nulls?
    .validate(result)
    .select('check', 'column', 'rule', 'status')
)
print(quality_check)
````

### âœ… ×¤×ª×¨×•×Ÿ

````python
# ×¤×ª×¨×•×Ÿ:

class PokemonValidator:
    """
    ××—×œ×§×” ×œ×‘×“×™×§×ª ××™×›×•×ª × ×ª×•× ×™ ×¤×•×§×™××•× ×™×
    """
    
    def __init__(self, df: pl.DataFrame):
        self.df = df
        self.results = []
    
    def check_completeness(self):
        """×‘×“×™×§×ª ×©×œ××•×ª ×¢××•×“×•×ª ×—×©×•×‘×•×ª"""
        check = Check(CheckLevel.WARNING, 'Completeness')
        important_cols = ['Name', 'Type 1', 'HP', 'Attack', 'Defense']
        result = (
            check
            .are_complete(important_cols)
            .validate(self.df)
        )
        self.results.append(result)
        return result
    
    def check_uniqueness(self):
        """×‘×“×™×§×ª ×™×™×—×•×“×™×•×ª ×©××•×ª"""
        check = Check(CheckLevel.WARNING, 'Uniqueness')
        result = (
            check
            .is_unique('Name')
            .validate(self.df)
        )
        self.results.append(result)
        return result
    
    def check_stats_positive(self):
        """×‘×“×™×§×” ×©×›×œ ×”×¡×˜×˜×™×¡×˜×™×§×•×ª ×—×™×•×‘×™×•×ª"""
        check = Check(CheckLevel.WARNING, 'Stats Validation')
        stats_cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
        for col in stats_cols:
            check = check.is_greater_than(col, 0)
        
        result = check.validate(self.df)
        self.results.append(result)
        return result
    
    def check_types(self):
        """×‘×“×™×§×ª ×˜×™×¤×•×¡×™× ××§×•×‘×œ×™×"""
        check = Check(CheckLevel.WARNING, 'Type Validation')
        accepted_types = [
            'Grass', 'Fire', 'Water', 'Bug', 'Normal', 'Poison',
            'Electric', 'Ground', 'Fairy', 'Fighting', 'Psychic',
            'Rock', 'Ghost', 'Ice', 'Dragon', 'Dark', 'Steel', 'Flying'
        ]
        result = (
            check
            .is_contained_in('Type 1', set(accepted_types))
            .validate(self.df)
        )
        self.results.append(result)
        return result
    
    def run_all_checks(self) -> pl.DataFrame:
        """××¨×™×¥ ××ª ×›×œ ×”×‘×“×™×§×•×ª ×•××—×–×™×¨ ×¡×™×›×•×"""
        print("ğŸ” ××¨×™×¥ ×‘×“×™×§×•×ª ××™×›×•×ª...\n")
        
        self.results = []  # ××™×¤×•×¡ ×ª×•×¦××•×ª
        
        # ×”×¨×¦×ª ×›×œ ×”×‘×“×™×§×•×ª
        self.check_completeness()
        self.check_uniqueness()
        self.check_stats_positive()
        self.check_types()
        
        # ××™×—×•×“ ×›×œ ×”×ª×•×¦××•×ª
        all_results = pl.concat(self.results)
        
        # ×¡×™×›×•×
        summary = all_results.select('check', 'column', 'rule', 'status', 'pass_rate')
        
        # ×”×“×¤×¡×ª ×¡×™×›×•×
        total_checks = len(summary)
        passed = summary.filter(pl.col('status') == 'PASS').shape[0]
        failed = total_checks - passed
        
        print(f"\n{'='*50}")
        print(f"âœ… ×¡×”\"×› ×‘×“×™×§×•×ª: {total_checks}")
        print(f"âœ… ×¢×‘×¨×•: {passed}")
        print(f"âŒ × ×›×©×œ×•: {failed}")
        print(f"ğŸ“Š ××—×•×– ×”×¦×œ×—×”: {passed/total_checks*100:.1f}%")
        print(f"{'='*50}\n")
        
        return summary

# ×©×™××•×©:
df = lf.collect()
validator = PokemonValidator(df)
results = validator.run_all_checks()
print(results)
````
