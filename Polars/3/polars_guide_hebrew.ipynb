````python
# ×¤×•× ×§×¦×™×” ×œ×™×¦×™×¨×ª ××¡×¤×¨ ×–×”×•×ª ××–×•×™×£
def get_random_nums(num_list, length):
    """×™×•×¦×¨ ××—×¨×•×–×ª ××§×¨××™×ª ×©×œ ×¡×¤×¨×•×ª"""
    return ''.join(str(n) for n in random.sample(num_list, length))

# ×™×¦×™×¨×ª SSN (Social Security Numbers) ××–×•×™×¤×™×
fake_ssns = []
nums = list(range(10))

random.seed(10)  # ×œ×©×—×–×•×¨ ×ª×•×¦××•×ª

for i in range(100):  # 100 ××¡×¤×¨×™× ××–×•×™×¤×™×
    part_1 = get_random_nums(nums, 3)
    part_2 = get_random_nums(nums, 2)
    part_3 = get_random_nums(nums, 4)
    fake_ssn = f'{part_1}-{part_2}-{part_3}'
    fake_ssns.append(fake_ssn)

# ×™×¦×™×¨×ª DataFrame
df_with_ssn = pl.DataFrame({'SSN': fake_ssns[:5], 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']})

print("ğŸ”‘ ×“×•×’×××•×ª SSN (××–×•×™×¤×™×):")
df_with_ssn
````

````python
# ×”×’×“×¨×ª ×§×‘×•×¦×•×ª ×’×™×œ
age_groups = [
    '0-17 years', '18-29 years', '30-39 years', 
    '40-49 years', '50-64 years', '65-74 years', 
    '75-84 years', '85 years and over', 'All Ages'
]

# ×¡×™× ×•×Ÿ ×•×˜×¢×™× ×ª × ×ª×•× ×™×
df = (
    pl.read_csv('../data/covid_19_deaths.csv')
    .filter(
        pl.col('Month').is_not_null(),
        pl.col('Age Group').is_in(age_groups)
    )
    .with_columns(
        pl.col('Data As Of').str.strptime(pl.Date, '%m/%d/%Y'),
        pl.col('Start Date').str.strptime(pl.Date, '%m/%d/%Y'),
        pl.col('End Date').str.strptime(pl.Date, '%m/%d/%Y'),
        pl.col('Year').cast(pl.Int64),
        pl.col('Month').cast(pl.Int64)
    )
)

print("âœ… ×”× ×ª×•× ×™× ××•×›× ×™× ×œ×•×™×–×•××œ×™×–×¦×™×”!")
print(f"ğŸ“Š ×¦×•×¨×”: {df.shape}")
````

````python
# ×”×›× ×ª × ×ª×•× ×™×: ××§×¨×™ ××•×•×ª ×œ×¤×™ ×§×‘×•×¦×•×ª ×’×™×œ
covid_deaths_by_age = (
    df
    .filter(
        pl.col('State') == 'United States',
        pl.col('Year') == 2023,
        pl.col('Age Group') != 'All Ages',
        pl.col('Sex') == 'All Sexes'
    )
    .group_by('Age Group')
    .agg(pl.col('COVID-19 Deaths').sum())
    .sort(by='COVID-19 Deaths', descending=True)
)

# ×™×¦×™×¨×ª ×ª×¨×©×™×
fig = px.bar(
    covid_deaths_by_age,
    x='Age Group',
    y='COVID-19 Deaths',
    title='××§×¨×™ ××•×•×ª ×-COVID-19 ×‘-2023 ×œ×¤×™ ×§×‘×•×¦×•×ª ×’×™×œ - ××¨×”"×‘',
    labels={'Age Group': '×§×‘×•×¦×ª ×’×™×œ', 'COVID-19 Deaths': '××¡×¤×¨ ××§×¨×™ ×”××•×•×ª'}
)

fig.update_layout(xaxis_title=None)
fig.show()

print("ğŸ’¡ ×˜×™×¤: ×”×¢×‘×¨ ×¢×›×‘×¨ ××¢×œ ×”×¢××•×“×•×ª ×œ×¨××•×ª × ×ª×•× ×™× ××“×•×™×§×™×!")
````

````python
# ××™×¤×•×™ ××“×™× ×•×ª ×œ××–×•×¨×™× (×œ×“×•×’××”)
us_state_division_dict = {
    'Alabama': 'East South Central',
    'California': 'Pacific',
    'Florida': 'South Atlantic',
    'Texas': 'West South Central',
    'New York': 'Mid-Atlantic'
    # ... × ×™×ª×Ÿ ×œ×”×•×¡×™×£ ×¢×•×“ ××“×™× ×•×ª
}

# ×”×›× ×ª × ×ª×•× ×™×
covid_vs_flu = (
    df
    .with_columns(
        pl.col('State')
        .replace_strict(us_state_division_dict, default='Others')
        .alias('Division')
    )
    .filter(
        pl.col('State') != 'United States',
        pl.col('Year') == 2023
    )
    .group_by('State', 'Division')
    .agg(
        pl.col('COVID-19 Deaths').sum(),
        pl.col('Influenza Deaths').sum(),
        pl.col('Pneumonia Deaths').sum()
    )
    .head(20)  # ×¨×§ 20 ×”×¨××©×•× ×•×ª ×œ×‘×”×™×¨×•×ª
)

# ×™×¦×™×¨×ª ×ª×¨×©×™×
fig = px.scatter(
    covid_vs_flu,
    x='COVID-19 Deaths',
    y='Influenza Deaths',
    color='Division',
    size='Pneumonia Deaths',
    hover_name='State',
    title='××§×¨×™ ××•×•×ª: COVID-19 ××•×œ ×©×¤×¢×ª - 2023',
    labels={
        'COVID-19 Deaths': '××§×¨×™ ××•×•×ª ×-COVID-19',
        'Influenza Deaths': '××§×¨×™ ××•×•×ª ××©×¤×¢×ª'
    }
)

fig.show()

print("ğŸ’¡ ×˜×™×¤: ×’×•×“×œ ×”× ×§×•×“×•×ª ××™×™×¦×’ ××§×¨×™ ××•×•×ª ××“×œ×§×ª ×¨×™××•×ª")
````

## 7ï¸âƒ£ ×–×™×”×•×™ ×•×˜×™×¤×•×œ ×‘×¢×¨×›×™× ×—×¨×™×’×™× (Outliers)

### ğŸ¯ ××” ×”× ×¢×¨×›×™× ×—×¨×™×’×™×?

×¢×¨×›×™× ×—×¨×™×’×™× (Outliers) ×”× **× ×§×•×“×•×ª × ×ª×•× ×™× ×©×—×•×¨×’×•×ª ××©××¢×•×ª×™×ª** ××”×ª×‘× ×™×ª ×”×›×œ×œ×™×ª:

- ğŸ“Š ×¢×œ×•×œ×™× ×œ×¢×•×•×ª × ×™×ª×•×—×™× ×¡×˜×˜×™×¡×˜×™×™×
- ğŸ¤– ××©×¤×™×¢×™× ×¢×œ ××•×“×œ×™× ×©×œ ×œ××™×“×ª ××›×•× ×”
- âš ï¸ ×¢×©×•×™×™× ×œ×”×¦×‘×™×¢ ×¢×œ ×©×’×™××•×ª ××• ×ª×•×¤×¢×•×ª ××™×•×—×“×•×ª

### ğŸ” ×©×ª×™ ×©×™×˜×•×ª ×¢×™×§×¨×™×•×ª ×œ×–×™×”×•×™

#### 1ï¸âƒ£ ×©×™×˜×ª IQR (Interquartile Range)
#### 2ï¸âƒ£ ×©×™×˜×ª Z-Score

### ğŸ“¦ ×©×™×˜×” 1: IQR (×”××œ×¦×” ×œ×¨×•×‘ ×”××§×¨×™×)

#### ğŸ“š ××” ×–×” IQR?

**IQR (Interquartile Range)** = Q3 - Q1

- **Q1** = ×¨×‘×¢×•×Ÿ ×¨××©×•×Ÿ (25%)
- **Q3** = ×¨×‘×¢×•×Ÿ ×©×œ×™×©×™ (75%)
- **IQR** = ×”×˜×•×•×— ×©××›×™×œ 50% ×”×××¦×¢×™×™× ×©×œ ×”× ×ª×•× ×™×

**×—×¨×™×’ ××**:
- `×¢×¨×š < Q1 - 1.5 Ã— IQR`
- `×¢×¨×š > Q3 + 1.5 Ã— IQR`

````python
# × ×©×ª××© ×‘× ×ª×•× ×™ Iris
iris_df = pl.from_pandas(plotly.data.iris())

print("ğŸ” ×‘×•×“×§×™× ×¢×¨×›×™× ×—×¨×™×’×™× ×‘-sepal_width...")
iris_df.head()
````

````python
# ×©×œ×‘ 1: ×—×™×©×•×‘ Q1, Q3, IQR
q1 = pl.col('sepal_width').quantile(0.25)
q3 = pl.col('sepal_width').quantile(0.75)
iqr = q3 - q1
threshold = 1.5

# ×©×œ×‘ 2: ×—×™×©×•×‘ ×’×‘×•×œ×•×ª
lower_limit = q1 - iqr * threshold
upper_limit = q3 + iqr * threshold

print("ğŸ“Š ×¤×¨××˜×¨×™× ×¡×˜×˜×™×¡×˜×™×™×:")
print(f"   Q1 (×¨×‘×¢×•×Ÿ ×¨××©×•×Ÿ): {iris_df.select(q1).item():.2f}")
print(f"   Q3 (×¨×‘×¢×•×Ÿ ×©×œ×™×©×™): {iris_df.select(q3).item():.2f}")
print(f"   IQR: {iris_df.select(iqr).item():.2f}")
print(f"   ×’×‘×•×œ ×ª×—×ª×•×Ÿ: {iris_df.select(lower_limit).item():.2f}")
print(f"   ×’×‘×•×œ ×¢×œ×™×•×Ÿ: {iris_df.select(upper_limit).item():.2f}")
````

````python
# ×©×œ×‘ 3: ×–×™×”×•×™ ×—×¨×™×’×™×
outliers_iqr = iris_df.filter(
    (pl.col('sepal_width') < lower_limit) | 
    (pl.col('sepal_width') > upper_limit)
)

print(f"ğŸš¨ × ××¦××• {outliers_iqr.shape[0]} ×¢×¨×›×™× ×—×¨×™×’×™×:")
outliers_iqr
````

#### ğŸ—‘ï¸ ××¤×©×¨×•×ª 1: ×”×¡×¨×ª ×—×¨×™×’×™×

````python
# ×™×¦×™×¨×ª ××¡×›×ª ×‘×•×œ×™×× ×™×ª
is_outlier = (
    (pl.col('sepal_width') < lower_limit) | 
    (pl.col('sepal_width') > upper_limit)
)

# ×”×¡×¨×” (×©××™×¨×ª ×¨×§ ×”×œ×-×—×¨×™×’×™×)
df_no_outliers = iris_df.filter(is_outlier.not_())

print(f"ğŸ“Š × ×ª×•× ×™× ×œ×¤× ×™: {iris_df.shape[0]} ×©×•×¨×•×ª")
print(f"âœ… × ×ª×•× ×™× ××—×¨×™: {df_no_outliers.shape[0]} ×©×•×¨×•×ª")
print(f"ğŸ—‘ï¸ ×”×•×¡×¨×•: {iris_df.shape[0] - df_no_outliers.shape[0]} ×©×•×¨×•×ª")

# ×•×™×“×•× ×©××™×Ÿ ×—×¨×™×’×™×
remaining_outliers = df_no_outliers.filter(is_outlier)
print(f"\nğŸ” ×—×¨×™×’×™× ×©× ×•×ª×¨×•: {remaining_outliers.shape[0]}")
````

#### ğŸ”„ ××¤×©×¨×•×ª 2: ×”×—×œ×¤×ª ×—×¨×™×’×™× ×‘×—×¦×™×•×Ÿ

````python
# ×”×—×œ×¤×” ×‘×—×¦×™×•×Ÿ ×‘××§×•× ×”×¡×¨×”
df_replaced = iris_df.with_columns(
    pl.when(is_outlier)
      .then(pl.col('sepal_width').median())  # ×× ×—×¨×™×’ -> ×—×¦×™×•×Ÿ
      .otherwise(pl.col('sepal_width'))      # ×× ×œ× -> ×¢×¨×š ××§×•×¨×™
      .alias('sepal_width')
)

print("ğŸ”„ ×”×—×œ×¤×ª ×—×¨×™×’×™× ×‘×—×¦×™×•×Ÿ:")
print(f"   ×—×¦×™×•×Ÿ: {iris_df.select('sepal_width').median().item():.2f}")

# ×‘×“×™×§×”: ×›××” ×©×•×¨×•×ª ×”×™×• ×—×¨×™×’×•×ª?
print(f"\nğŸ“Š ×©×•×¨×•×ª ×©×”×•×—×œ×¤×•: {outliers_iqr.shape[0]}")
df_replaced.filter(is_outlier).head()  # ×¦×¨×™×š ×œ×”×™×•×ª ×¨×™×§ ××• ×¢× ×¢×¨×›×™ ×—×¦×™×•×Ÿ
````

### ğŸ“Š ×©×™×˜×” 2: Z-Score (×œ×”×ª×¤×œ×’×•×ª × ×•×¨××œ×™×ª)

#### ğŸ“š ××” ×–×” Z-Score?

**Z-Score** = ××¡×¤×¨ ×¡×˜×™×•×ª ×ª×§×Ÿ ××”×××•×¦×¢

$$Z = \frac{x - \mu}{\sigma}$$

×›××©×¨:
- x = ×”×¢×¨×š
- Î¼ = ×××•×¦×¢
- Ïƒ = ×¡×˜×™×™×ª ×ª×§×Ÿ

**×—×¨×™×’ ××**: |Z-Score| > 3 (××• 2 ×œ××—××™×¨×™×)

````python
# ×—×™×©×•×‘ Z-Score
df_zscore = iris_df.with_columns(
    (
        (pl.col('sepal_width') - pl.col('sepal_width').mean()) / 
        pl.col('sepal_width').std()
    ).alias('sepal_width_zscore')
)

print("ğŸ“Š 5 ×”×¢×¨×›×™× ×¢× Z-Score ×”×’×‘×•×” ×‘×™×•×ª×¨:")
df_zscore.sort('sepal_width_zscore', descending=True).head()
````

````python
# ×–×™×”×•×™ ×—×¨×™×’×™× (|Z| > 3)
is_outlier_z = (
    (pl.col('sepal_width_zscore') > 3) | 
    (pl.col('sepal_width_zscore') < -3)
)

outliers_z = df_zscore.filter(is_outlier_z)

print(f"ğŸš¨ ×—×¨×™×’×™× ×œ×¤×™ Z-Score: {outliers_z.shape[0]}")
outliers_z
````

````python
# ×”×¡×¨×ª ×—×¨×™×’×™×
df_z_cleaned = df_zscore.filter(is_outlier_z.not_())

print(f"ğŸ“Š ×œ×¤× ×™: {df_zscore.shape[0]} ×©×•×¨×•×ª")
print(f"âœ… ××—×¨×™: {df_z_cleaned.shape[0]} ×©×•×¨×•×ª")
print(f"ğŸ—‘ï¸ ×”×•×¡×¨×•: {df_zscore.shape[0] - df_z_cleaned.shape[0]} ×©×•×¨×•×ª")
````
