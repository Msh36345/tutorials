````python
# ×§×¨×™××” ×¢×¦×œ×” ×©×œ CSV
print("×§×¨×™××” ×¢×¦×œ×” ×©×œ ×§×•×‘×¥ ×’×“×•×œ:")

# ×‘××§×•× pl.read_csv() × ×©×ª××© ×‘-scan_csv()
lazy_df = pl.scan_csv('../data/titanic_dataset.csv')

# ×‘×™×¦×•×¢ ×¤×¢×•×œ×•×ª - ×¢×“×™×™×Ÿ ×œ× ×§×•×¨× ××ª ×›×œ ×”×§×•×‘×¥!
result = (
    lazy_df
    .filter(pl.col('Age') > 30)
    .select(['Name', 'Age', 'Fare'])
    .collect()  # ×¨×§ ×›××Ÿ ×§×•×¨× ××ª ××” ×©×¦×¨×™×š
)

print(result.head())
print("\nğŸ’¡ Polars ×§×¨× ×¨×§ ××ª ×”×¢××•×“×•×ª ×•×”×©×•×¨×•×ª ×”× ×“×¨×©×•×ª!")
````

### 2ï¸âƒ£ Streaming Mode - ×¢×™×‘×•×“ ×‘×–×¨×™××”


````python
# Streaming - ×œ×§×‘×¦×™× ×××© ×’×“×•×œ×™×
print("Streaming mode:")

result = (
    pl.scan_csv('../data/titanic_dataset.csv')
    .filter(pl.col('Survived') == 1)
    .group_by('Sex').agg([
        pl.count().alias('count'),
        pl.col('Age').mean().alias('avg_age')
    ])
    .collect(streaming=True)  # ××¤×¢×™×œ streaming
)

print(result)
print("\nâœ¨ Streaming ××¢×‘×“ ×‘×—×œ×§×™× ×œ×œ× ×˜×¢×™× ×ª ×›×œ ×”×§×•×‘×¥!")
````
